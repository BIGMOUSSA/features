## CONVERITNG MOBILE DATA INTO CIDER FORMAT

```python
#read data
data = pd.read_csv("../data/data_mobile/data.csv", index_col = 0)
```

```python
data.head()
```

```python
data["time_lengh"] = data["call_time"].astype(str).apply(lambda x : len(x))
```

```python
data[data["time_lengh"]<=4]
```

```python
#caller_id
#volume
#timestamp
data.rename(columns = {"msisdn":"caller_id"}, inplace = True)
```

```python
from to_timestamp import to_timestamp
```

```python
data_out = to_timestamp(data=data, date_col="call_date", time_col="call_time")
```

```python
data_out[data_out["time_lengh"]<=4]
```

```python
data_out.drop(columns = ["call_date", "call_time"], inplace = True)
```

```python
data_out.to_csv("../data/mobiledata_cider.csv")
```

## CONVERTING TRANSACTION (MOBILE MONEY) DATA IN CIDER FORMAT

```python
transaction = pd.read_csv("../data/transactions/om.csv", index_col=0)
```

```python
transaction.direction.value_counts()
```

```python
transaction
```

```python
transaction = to_timestamp(transaction, date_col="transfer_date", time_col="transfer_time")
```

```python
transaction.rename(columns={"transaction_amount" : "amount", "msisdn":"caller_id"}, inplace=True)
```

```python
transaction.drop(columns=["transfer_date", "transfer_time"], inplace=True)
```

```python
transaction
```

```python
transaction["region"] = transaction["localite"].astype(str).apply(lambda x : x.split("/")[-1])
```
